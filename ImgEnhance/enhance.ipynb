```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image (grayscale or color)
image_path = 'xray.jpg'  # Replace with your image path
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)  # For grayscale, use IMREAD_GRAYSCALE

# Original Image
plt.subplot(2, 3, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

# Histogram Equalization
hist_eq_image = cv2.equalizeHist(image)
plt.subplot(2, 3, 2)
plt.imshow(hist_eq_image, cmap='gray')
plt.title('Histogram Equalization')
plt.axis('off')

# CLAHE (Contrast Limited Adaptive Histogram Equalization)
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
clahe_image = clahe.apply(image)
plt.subplot(2, 3, 3)
plt.imshow(clahe_image, cmap='gray')
plt.title('CLAHE')
plt.axis('off')

# Unsharp Masking (for edge enhancement)
blurred_image = cv2.GaussianBlur(image, (9, 9), 10.0)
unsharp_image = cv2.addWeighted(image, 1.5, blurred_image, -0.5, 0)
plt.subplot(2, 3, 4)
plt.imshow(unsharp_image, cmap='gray')
plt.title('Unsharp Masking')
plt.axis('off')

# Noise Reduction (using a bilateral filter)
denoised_image = cv2.bilateralFilter(image, d=9, sigmaColor=75, sigmaSpace=75)
plt.subplot(2, 3, 5)
plt.imshow(denoised_image, cmap='gray')
plt.title('Noise Reduction')
plt.axis('off')

# Edge Detection (using Canny edge detection)
edges_image = cv2.Canny(image, 100, 200)
plt.subplot(2, 3, 6)
plt.imshow(edges_image, cmap='gray')
plt.title('Edge Detection')
plt.axis('off')

# Display the results
plt.tight_layout()
plt.show()
```
